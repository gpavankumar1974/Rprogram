```r
#Installing the Psych package and loading it
#install.packages("psych")
library(psych)
#Loading the dataset
bfi_data=bfi
```

```r
head(bfi_data)
# The columns are already classified into 5 factors thus their names 
# start with letters A (Agreeableness), C (Conscientiousness), E (Extraversion), N (Neuroticism) and O (Openness). 

colnames(bfi_data)
```

```r
bfi_data=bfi_data[complete.cases(bfi_data),]
```

```r
#Create the correlation matrix from bfi_data
bfi_cor <- cor(bfi_data)
bfi_cor
```

```r
#Factor analysis of the data
library("GPArotation")
factors_data <- fa(r = bfi_cor, nfactors = 6)
#Getting the factor loadings and model analysis
factors_data
#install.packages("GPArotation")

# minimum residua = minres
# fa , uls, ols etc. methods
```

```r
# The factor loadings show that the first factor represents N followed by C,E,A and O.
#This means most of the members in the data have Neuroticism in the data. 
```

```r
parallel <- fa.parallel(bfi_cor, fm = 'minres', fa = 'fa')
```

```r
threefactor <- fa(bfi_cor,nfactors = 6,rotate = "oblimin",fm="minres")
#print(threefactor)
print(threefactor$loadings,cutoff = 0.3)
```

```r
fa.diagram(threefactor)
```
